#  prioritize the detection of emotions that are present in your dataset.

```python
def classify_emotions_prioritized_with_scores(text):
    emotion_obj = NRCLex(text)
    emotion_scores = {emotion: 0 for emotion in dataset_emotions}  # Initialize scores for dataset emotions
    
    # Update scores based on NRCLex analysis
    for emotion in dataset_emotions:
        emotion_scores[emotion] = emotion_obj.affect_frequencies.get(emotion, 0)
    
    # Identify the primary emotion based on the highest score
    primary_emotion = max(emotion_scores, key=emotion_scores.get)
    primary_score = emotion_scores[primary_emotion]
    
    if primary_score == 0:  # If the highest score is 0, no dataset emotion is detected
        return 'unknown', emotion_scores
    else:
        return primary_emotion, emotion_scores
```

```python
def preprocess_text(text):
    # Tokenization
    tokens = word_tokenize(text)
    
    # Lowercasing
    tokens = [word.lower() for word in tokens]
    
    # Remove numbers
    tokens = [word for word in tokens if not word.isdigit()]
    
    # Remove punctuation and symbols
    tokens = [word for word in tokens if word.isalpha()]
    
    # Stopwords removal
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if not word in stop_words]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    # Re-join tokens into a single string
    return ' '.join(tokens)
```
