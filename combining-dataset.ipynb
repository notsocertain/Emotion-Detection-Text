```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df1= pd.read_csv('dataset-1/train.txt', names=['text','emotion'], delimiter=';')
```

```python
df2= pd.read_csv('dataset-1/test.txt', names=['text','emotion'], delimiter=';')
```

```python
df3= pd.read_csv('dataset-1/val.txt', names=['text','emotion'], delimiter=';')
```

```python
data1 = pd.concat((df1,df2,df3), axis=0)
data1
```

```python
data1.shape
```

```python
# classes and value counts:
print(data1.emotion.value_counts())
print("Number of classes: ",len(data1.emotion.value_counts()))
```

```python
# Getting the emotion counts dynamically
emotion_counts = data1['emotion'].value_counts()

# Plotting the pie chart dynamically
plt.figure(figsize=(10, 7))
plt.pie(emotion_counts, labels=emotion_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Emotion Distribution in first dataset')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.show()
```

# MERGING data1 and data2

```python
# merge data from both the sources into 1 dataset:
data = pd.concat((data1,data2),axis=0)
data
```

```python
# Getting the emotion counts dynamically
emotion_counts = data['emotion'].value_counts()

# Plotting the pie chart dynamically
plt.figure(figsize=(10, 7))
plt.pie(emotion_counts, labels=emotion_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Emotion Distribution in  dataset')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.show()
```
